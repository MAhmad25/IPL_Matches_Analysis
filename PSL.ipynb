```python
#Which team won the most matches
match_wins=ipl["match_winner"].value_counts()
df = match_wins.reset_index()
df.columns = ["team", "wins"]
plx.bar(df,title="Matches Win",x="team",y="wins")
```

```python
#Toss Decision Trend
plx.histogram(ipl["toss_decision"])
```

```python
#Toss Winner vs Match Winner
countOfTossWinMatchWin=ipl[ipl["toss_winner"]==ipl["match_winner"]]["match_id"].count()
pecentage=(countOfTossWinMatchWin/ipl.shape[0])*100
print(pecentage.round(2))
```

```python
#How do team wins( Run vs wickets )
ipl["won_by"].value_counts()
plx.histogram(ipl["won_by"],title="WON By")
```

<h1>Key Player Performance</h1>

```python
# 1 Most "Player of the Match" Awards
playerOfTheMatchCount=ipl["player_of_the_match"].value_counts()
df=playerOfTheMatchCount.reset_index()
df.columns=["Player Name","Count"]
plx.bar(df,x="Player Name",y="Count")
```

```python
# 2 Top Scorer
top2=ipl.groupby("top_scorer")["highscore"].sum().nlargest(2)
plx.bar(top2,title="High Scorer")
```

```python
#10 Best Bowling Figure
# First we will extract the wicket from the best_bowling_figure
ipl["highest_wicket"]=ipl["best_bowling_figure"].apply(lambda x:x.split("--")[0])
ipl["highest_wicket"]=ipl["highest_wicket"].astype(int)
best10=ipl.groupby("best_bowling")["highest_wicket"].sum().sort_values(ascending=False).head(10)
plx.bar(best10)
```

```python
#Most Matches Played by Venue
matchesPlayedByVenue=ipl["venue"].value_counts()
plx.bar(matchesPlayedByVenue)
```

```python
#Who won the highest margin by run
whoWon=ipl.groupby("won_by")["margin"].max()
ipl[ipl["margin"]==whoWon.iloc[0]][["match_winner","margin"]]
```

```python
#Which player have the highest individual Score
highScorer=ipl.groupby("top_scorer")["highscore"].max().nlargest(1)
pd.DataFrame(highScorer)
```

```python
highBowler=ipl.groupby("best_bowling")["highest_wicket"].max().nlargest(1)
pd.DataFrame(highBowler)
```
